```python
"""Run Gdrive mount tests"""
import os

folder_path = '/content/drive/MyDrive/project/AEGIS/data'

# Check if the directory exists
if os.path.exists(folder_path):
    # List all entries in the directory
    entries = os.listdir(folder_path)

    # Filter out only directories
    subfolders = [entry for entry in entries if os.path.isdir(os.path.join(folder_path, entry))]

    if subfolders:
        print(f"'{folder_path}' 아래의 하위 폴더 목록:")
        for folder in subfolders:
            print(f"{folder}")
    else:
        print(f"'{folder_path}' 아래에 하위 폴더가 없습니다.")
else:
    print(f"경로 '{folder_path}'를 찾을 수 없습니다.")
```

```python
import os

folder_path = '/content/drive/MyDrive/project/AEGIS/Dataset/burglary/실내'

# Check if the directory exists
if os.path.exists(folder_path):
    # List all entries in the directory
    entries = os.listdir(folder_path)

    # Filter out only files
    files_in_folder = [entry for entry in entries if os.path.isfile(os.path.join(folder_path, entry))]

    if files_in_folder:
        # sorted_files = sorted(files_in_folder, reverse=True)
        sorted_files = sorted(files_in_folder)

        print(f"'{folder_path}' 아래의 파일 목록:")
        print(f"{len(files_in_folder)}개")

        # for file_name in files_in_folder:
        for file_name in sorted_files:
            print(f"{file_name}")
    else:
        print(f"'{folder_path}' 아래에 파일이 없습니다.")
else:
    print(f"경로 '{folder_path}'를 찾을 수 없습니다.")
```
